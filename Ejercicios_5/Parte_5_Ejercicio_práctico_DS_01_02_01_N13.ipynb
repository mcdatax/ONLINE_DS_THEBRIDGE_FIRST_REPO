## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]
# Convertir todas las palabras a minúsculas para evitar problemas de mayúsculas/minúsculas al buscar duplicados
ejer_3 = [palabra.lower() for palabra in ejer_3]

# Encontrar palabras duplicadas en la lista
def encontrar_duplicados(ejer_3):
    conteos = {} # Con esta variable contaremos cuántas veces aparece cada palabra en la lista
    for palabra in ejer_3: # Iteramos sobre cada palabra en la lista
        '''
        El método get() busca la clave palabra en el diccionario conteos. 
        Si existe, devuelve su valor actual; si no, devuelve el valor por defecto 0.
        Esto evita errores si la palabra no está en el diccionario.
        Luego suma 1 al conteo y lo asigna de vuelta a conteos[palabra] para actualizar el conteo de esa palabra.
        '''
        conteos[palabra] = conteos.get(palabra, 0) + 1 # Si la palabra ya está en el diccionario, incrementamos su conteo, si no, la agregamos con un conteo de 1
    # print(conteos) # Imprimimos el diccionario de conteos para verificar que se están contando correctamente las palabras
    
    duplicados = {}
    
    for palabra, count in conteos.items():
        if count > 1:
            '''
            Función sin compresion de listas para encontrar los índices de todas las ocurrencias de la palabra duplicada en la lista ejer_3
            indices = []
            for i, p in enumerate(ejer_3):
                if p == palabra:
                indices.append(i)
            '''
            # Usamos una comprensión de listas para encontrar los índices de todas las ocurrencias de la palabra duplicada en la lista ejer_3
            indices = [i for i, p in enumerate(ejer_3) if p == palabra]
            # Agregamos la palabra duplicada y sus índices al diccionario de duplicados
            duplicados[palabra] = indices
    return duplicados

# Ejemplo de uso
duplicados_encontrados = encontrar_duplicados(ejer_3)
for palabra, indices in duplicados_encontrados.items():
    print(f"La palabra '{palabra}' se encuentra en los índices: {indices}")

`````

## Ejercicio 10
Convierte el programa del ejercicio 2 en una función al que se le pase el valor a buscar como un argumento posicional

`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]

def buscar_duplicados(palabra_buscar):
    """
    Busca todas las posiciones de una palabra específica en la lista ejer_3
    
    Args:
        palabra_buscar: La palabra a buscar (argumento posicional)
    
    Returns:
        Lista con los índices donde aparece la palabra
    """
    # Convertir la lista a minúsculas
    lista_lower = [palabra.lower() for palabra in ejer_3]
    
    # Buscar los índices donde aparece la palabra
    indices = [i for i, p in enumerate(lista_lower) if p == palabra_buscar.lower()]
    
    if indices:
        print(f"La palabra '{palabra_buscar}' se encuentra en los índices: {indices}")
    else:
        print(f"La palabra '{palabra_buscar}' no se encontró en la lista.")
    
    return indices

# Pruebas con las palabras
buscar_duplicados("un")
buscar_duplicados("es")
buscar_duplicados("binario")
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
DNI= input("Introduce tu DNI: ")
palabra='TRWAGMYFPDXBNJZSQVHLCKE'

def validar_dni(dni):
    #Eliminar espacios en blanco y convertir a mayúsculas
    dni = str(dni).strip().upper()

    # Verificar longitud (debe ser entgre 8 y 9 caracteres: 7 - 8 dígitos + 1 letra)
    if len(dni) < 8 or len(dni) > 9:
        print("DNI no válido: longitud incorrecta.")
        return False
    
    # Separar número y letra
    numero = dni[:-1]  # Todos los caracteres excepto el último que es la letra
    letra = dni[-1]    # El último carácter es la letra

    # Verificar que los número sea un dígito
    if not numero.isdigit():
        print("DNI no válido: los primeros caracteres dçeben ser dígitos.")
        return False
    
    # Verificar que tenga 7 u 8 dígitos
    if len(numero) < 7 or len(numero) > 8:
        print("DNI no válido: el número debe tener entre 7 y 8 dígitos.")
        return False

    # Verificar que el ultimo carácter sea una letra
    if not letra.isalpha():
        print("DNI no válido: el último carácter debe ser una letra.")
        return False

    # Calcular la letra correcta
    palabra = 'TRWAGMYFPDXBNJZSQVHLCKE'
    letra_calculada = palabra[int(numero) % 23]

    # Verificar si la letra calculada coincide con la letra ingresada
    if letra_calculada != letra:
        print(f"DNI no válido: la letra correcta para el número {numero} es '{letra_calculada}', no '{letra}'.")
        return False

    print(f"DNI válido: {dni}")
    return True


validar_dni(DNI)
`````
