## Ejercicio 1
Para este ejercicio vamos a poner en práctica [las funciones built in](https://docs.python.org/3/library/functions.html).

1. Calcula el máximo de la lista: [4, 6, 8, -1]
2. Suma todos los elementos de la lista anterior
3. Redondea este float a 3 dígitos decimales: 63.451256965
4. Valor absoluto de: -74

```python
lista = [4,6,8,-1]
print(f"El número máximo de la lista es: {max(lista)}")
print(f"La suma de la lista es: {sum(lista)}")
decimal = 63.451256965
print(f"El número redondeado a 3 decimales es: {round(decimal, 3)}")
numero = -74
print(f"El valor absoluto de {numero} es: {abs(numero)}")
```

## Ejercicio 2
Para el siguiente string se pide imprimir por pantalla los siguientes casos

    "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."

1. Pásalo todo a mayúsculas
2. Pásalo todo a minúsculas
3. Solo la primera letra de cada palabra en mayúscula, el resto en minúscula
4. Crea una lista dividiéndolo por sus espacios
5. Sustituye las comas `,` por puntos y comas `;`
6. Elimina las `a` minúsculas

```python
cadena = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."
print(f"Texto en mayúsculas: {cadena.upper()}")
print(f"Texto en minúsculas: {cadena.lower()}")
print(f"Texto con la primera letra de cada palabra en mayúscula: {cadena.title()}")
lista = cadena.split()
print(f"Lista de palabras: {lista}")
cadena_reemplazada = cadena.replace(",", ";")
print(f"Cadena con comas reemplazadas por punto y coma: {cadena_reemplazada}")
cadena_sin_a_minuscula = cadena.replace("a", "")
print(f"Cadena sin la letra 'a' minúscula: {cadena_sin_a_minuscula}")
```

## Ejercicio 3
1. Crea una lista con 3 elementos numéricos
2. Añade un cuarto elemento
3. Calcula la suma de todos
4. Elimina el segundo elemento de la lista
5. Añade otro elemento en la posicion 3 de la lista
6. Crea otra lista con 4 elementos y concaténala a la que ya tenías.
7. Ordena la lista de menor a mayor.
8. BONUS: Ordena la lista de mayor a menor.  
(*PISTA: Para hacer 8, tendrás que buscar en la información del método...*)  
  
  
NOTA: Pinta por pantalla el resultado de cada paso.

```python
lista = [4,6,8]
lista.append(10)
print(f"Lista después de agregar un nuevo número: {lista}")
print(f"La suma de los elementos es: {sum(lista)}")
lista.pop(1)
print(f"Lista después de eliminar el número 1: {lista}")
lista.insert(2, 12)
print(f"Lista después de insertar el número 12 en la posición 3: {lista}")
lista2 = [9,33,567,342]
lista.extend(lista2)
print(f"Lista después de extenderla con otra lista: {lista}")
lista.sort()
print(f"Lista ordenada de menor a mayor: {lista}")
lista.sort(reverse=True)
print(f"Lista ordenada de mayor a menor: {lista}")
```
