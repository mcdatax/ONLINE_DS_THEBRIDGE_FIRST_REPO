```python
lista = [4,6,8,-1]
print(f"El número máximo de la lista es: {max(lista)}")
print(f"La suma de la lista es: {sum(lista)}")
decimal = 63.451256965
print(f"El número redondeado a 3 decimales es: {round(decimal, 3)}")
numero = -74
print(f"El valor absoluto de {numero} es: {abs(numero)}")
```

```python
cadena = "En un lugar de la Mancha, de cuyo nombre no quiero acordarme."
print(f"Texto en mayúsculas: {cadena.upper()}")
print(f"Texto en minúsculas: {cadena.lower()}")
print(f"Texto con la primera letra de cada palabra en mayúscula: {cadena.title()}")
lista = cadena.split()
print(f"Lista de palabras: {lista}")
cadena_reemplazada = cadena.replace(",", ";")
print(f"Cadena con comas reemplazadas por punto y coma: {cadena_reemplazada}")
cadena_sin_a_minuscula = cadena.replace("a", "")
print(f"Cadena sin la letra 'a' minúscula: {cadena_sin_a_minuscula}")
```

```python
lista = [4,6,8]
lista.append(10)
print(f"Lista después de agregar un nuevo número: {lista}")
print(f"La suma de los elementos es: {sum(lista)}")
lista.pop(1)
print(f"Lista después de eliminar el número 1: {lista}")
lista.insert(2, 12)
print(f"Lista después de insertar el número 12 en la posición 3: {lista}")
lista2 = [9,33,567,342]
lista.extend(lista2)
print(f"Lista después de extenderla con otra lista: {lista}")
lista.sort()
print(f"Lista ordenada de menor a mayor: {lista}")
lista.sort(reverse=True)
print(f"Lista ordenada de mayor a menor: {lista}")
```
