```python
import time  # Para delays y animación
precio_pizza_familiar = 8.95
precio_pizza_mediana = 9.9

numero_pizzas_familiar = int(input("Buenas tardes.\n"
                                   "Bienvenido al servicio de pedido online.\n"
                                   "El precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana\n"
                                   "¿Cuántas pizzas familiares desea? "))

numero_pizzas_mediana = int(input("Estupendo, ¿cuántas pizzas medianas desea? "))
print(f"Estupendo, se están preparando {numero_pizzas_familiar} pizzas familiares y {numero_pizzas_mediana} pizzas medianas. Dígame su dirección")
direccion = input("Dirección: ")

total_pizzas = numero_pizzas_familiar + numero_pizzas_mediana
precio_total = (numero_pizzas_familiar * precio_pizza_familiar) + (numero_pizzas_mediana * precio_pizza_mediana)

# Animación de "cargando" (cursor giratorio)
print(" Preparando las pizzas...", end="")
for _ in range(10):  # 10 iteraciones
    for char in "|/-\\":  # Iconos del cursor
        print(char, end="\r", flush=True)  # \r regresa al inicio de la línea
        time.sleep(0.2)  # Pausa de 0.2 segundos
print(" ¡Listo!")  # Borra el cursor y muestra mensaje

print(f"Le mandaremos las {total_pizzas} pizzas a la dirección '{direccion}'. Serán {precio_total:.2f} euros. Muchas gracias por su pedido.")
```

```python
numeros = input("Ingresa 3 números separados por comas: ")
numeros_lista = [int(num.strip()) for num in numeros.split(",")]
print(f"El número máximo es: {max(numeros_lista)}")
numero_maximo = max(numeros_lista)
numero_minimo = min(numeros_lista)


print("\nVamos a determinar la verdad o falsedad de la siguiente afirmación: El número máximo menos el número mínimo es igual a la suma total dividida por 5")
time.sleep(2)  # Pausa de 2 segundos para crear suspenso
if (numero_maximo - numero_minimo) == (sum(numeros_lista)/5):
    print("\n\nLa afirmación es VERDADERA: El número máximo menos el número mínimo es IGUAL a la suma total dividida por 5")
else:
    print("\n\nLa afirmación es FALSA: El número máximo menos el número mínimo NO es IGUAL a la suma total dividida por 5")

print(f"\nJustificamos la afirmación con los cálculos: ")
print(f"El número máximo es: {numero_maximo}")
print(f"El número mínimo es: {numero_minimo}")
print(f"El máximo - el mínimo es: {numero_maximo - numero_minimo}")
print(f"La suma total es: {sum(numeros_lista)}")
print(f"La suma total dividida por 5 es: {sum(numeros_lista)/5}")
```
