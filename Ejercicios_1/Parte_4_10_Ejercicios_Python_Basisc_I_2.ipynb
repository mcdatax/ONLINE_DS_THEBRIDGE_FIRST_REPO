## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
# !pip install numpy # Instala la biblioteca NumPy, que es una biblioteca de Python para el cálculo numérico y el manejo de arrays.
import numpy as np # Importa la biblioteca NumPy y le asigna el alias "np" para facilitar su uso en el código. NumPy proporciona funciones y herramientas para trabajar con arrays, realizar operaciones matemáticas y científicas, entre otras funcionalidades y está hecho en C, lo que lo hace muy eficiente para cálculos numéricos.

grados = float(input("Ingresa el valor en grados para convertir a Radianes: "))
radianes = grados * (np.pi / 180)
print(f"{grados} grados son {radianes} radianes.")
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
base = float(input("Ingresa la base del paralelogramo: "))
altura = float(input("Ingresa la altura del paralelogramo: "))
area = base * altura
print(f"El área del paralelogramo es: {area}")
```

## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
# Reemplaza el contenido de la celda del Ejercicio 4 con esto:

base_str = input("Ingresa la base del paralelogramo (se debe usar coma para decimales, ej. 5,5): ")
base = float(base_str.replace(',', '.'))

altura_str = input("Ingresa la altura del paralelogramo (se debe usar coma para decimales, ej. 3,2): ")
altura = float(altura_str.replace(',', '.'))

area = base * altura
print(f"El área del paralelogramo es: {area}")
```
